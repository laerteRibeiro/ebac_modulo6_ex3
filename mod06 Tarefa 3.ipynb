#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esse arquivo.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
# %matplotlib notebook
```

```python
df = pd.read_csv('sinasc_RO_2019.csv')
df
```

```python
df.info()
```

```python
var_interesse = ['APGAR5', 'SEXO', 'GESTACAO', 'PESO']
sinasc = df[var_interesse]
sinasc
```

```python
sinasc.dropna(inplace=True)
sinasc
```

```python
sinasc['GESTACAO'].replace('Menos de 22 semanas', '0 a 21 semanas', inplace=True)
sinasc
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
sinasc.loc[sinasc['APGAR5'] <= 3.0, 'CAT_APGAR5'] = 'asfixia grave'
sinasc.loc[(sinasc['APGAR5'] >= 4.0) & (sinasc['APGAR5'] <= 5.0), 'CAT_APGAR5'] = 'asfixia moderada'
sinasc.loc[(sinasc['APGAR5'] >= 6.0) & (sinasc['APGAR5'] <= 7.0), 'CAT_APGAR5'] = 'asfixia leve'
sinasc.loc[(sinasc['APGAR5'] >= 8.0) & (sinasc['APGAR5'] <= 10.0), 'CAT_APGAR5'] = 'normal'
sinasc
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.barplot(x='APGAR5'
               , y='GESTACAO'
               , data=sinasc
               , hue= 'SEXO'
               , ci=None
               )
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
freq = pd.crosstab(sinasc['APGAR5'], sinasc['GESTACAO'])
freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq_pct_linha = freq.div(freq.sum(axis=1), axis=0)
freq_pct_linha
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
freq_pct_linha.plot.bar(figsize=(18,8))
```

Pelos dados apresentados no gráfico, eu concordo com a afirmação dos médicos. 
Além de ser a minoria, os partos realizados com menos de 37 semanas apresentam maior incidência de asfixias moderadas e graves, o que pode indicar a formação não completa dos bebês.
